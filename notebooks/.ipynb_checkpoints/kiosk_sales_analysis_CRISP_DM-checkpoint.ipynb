# Kiosk Sales Analysis Using CRISP-DM

## Introduction
This project aims to analyze sales data from a kiosk that sells various items such as candies, drinks, cigarettes, and other products. By applying the CRISP-DM (Cross Industry Standard Process for Data Mining) methodology, we will systematically approach the data analysis process to extract valuable insights that can help increase sales and improve overall store performance.

The CRISP-DM methodology consists of six phases: Business Understanding, Data Understanding, Data Preparation, Modeling, Evaluation, and Deployment. Each phase is essential for ensuring that the data analysis is comprehensive and the results are actionable.

## 1. Business Understanding
In the Business Understanding phase, we aim to gain a clear understanding of the business context, objectives, and requirements. This involves defining the project goals, identifying the key stakeholders, and understanding how the data can be leveraged to address business challenges and opportunities.

### 1.1 Project Objectives
The primary goal of this project is to identify patterns and trends in sales data to provide data-driven recommendations for increasing sales. Specifically, we aim to:

1. **Understand the Sales Performance of Different Products and Categories**: Determine which products and categories are performing well and which are not.
2. **Identify the Most and Least Popular Items**: Recognize which items have the highest and lowest sales volumes.
3. **Analyze Sales Trends Over Time**: Investigate any trends or seasonal patterns in sales data.
4. **Evaluate the Impact of Unit Price on the Quantity Sold**: Assess how changes in unit price affect the quantity sold of various products.
5. **Develop a Predictive Model to Forecast Future Sales**: Create a model to predict future sales and guide inventory management.

### 1.2 Business Context

The kiosk operates in a competitive retail environment, offering a diverse range of products. The sales data provides valuable insights into customer preferences, purchasing behavior, and product performance. Understanding these aspects can help in making informed decisions to improve profitability and customer satisfaction.

#### 2.6.3 Sales Trends Over Time
We will examine sales trends over time to identify any patterns or seasonality in the sales data.

```python
df['Date Sale'] = pd.to_datetime(df['Date Sale'])
df['Day'] = df['Date Sale'].dt.day
df['Month'] = df['Date Sale'].dt.month
sales_by_day_month = df.groupby(['Month', 'Day'])['Total Sale'].sum().reset_index()
sales_by_day_month['Date'] = sales_by_day_month.apply(lambda row: f"{int(row['Month']):02d}-{int(row['Day']):02d}", axis=1)
plt.figure(figsize=(14, 7))
plt.plot(sales_by_day_month['Date'], sales_by_day_month['Total Sale'], marker='o')
plt.xticks(rotation=90)  # Rotar las etiquetas del eje x para mejor visibilidad
plt.title('Total Sales Over Time by Day and Month')
plt.xlabel('Date (Month-Day)')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

### 4.3 Model Selection and Training
We will start by selecting a few different modeling techniques and training them on our data. Common choices for sales prediction include linear regression, decision trees, and random forests.

```python
# Linear Regression Model
from sklearn.linear_model import LinearRegression

# Initialize the linear regression model
linear_model = LinearRegression()

# Train the model on the training data
linear_model.fit(x_train, y_train)
```

```python
# Decision Tree Model
from sklearn.tree import DecisionTreeRegressor

# Initialize the decision tree model
tree_model = DecisionTreeRegressor(random_state=42)

# Train the model on the training data
tree_model.fit(x_train, y_train)
```

```python
# Random Forest Model
from sklearn.ensemble import RandomForestRegressor

# Initialize the random forest model
forest_model = RandomForestRegressor(random_state=42)

# Train the model on the training data
forest_model.fit(x_train, y_train)
```

### 4.4 Model Evaluation
After training the models, we will evaluate their performance using the testing set. We will use metrics such as Mean Absolute Error (MAE), Mean Squared Error (MSE), and R-squared to assess the models.

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Function to evaluate the model
def evaluate_model(model, x_test, y_test):
    predictions = model.predict(x_test)
    mae = mean_absolute_error(y_test, predictions)
    mse = mean_squared_error(y_test, predictions)
    r2 = r2_score(y_test, predictions)
    return mae, mse, r2
```

```python
# Evaluate the linear regression model
linear_mae, linear_mse, linear_r2 = evaluate_model(linear_model, x_test, y_test)
f"Linear Regression - MAE: {linear_mae}, MSE: {linear_mse}, R2: {linear_r2}"
```

```python
# Evaluate the decision tree model
tree_mae, tree_mse, tree_r2 = evaluate_model(tree_model, x_test, y_test)
f"Decision Tree - MAE: {tree_mae}, MSE: {tree_mse}, R2: {tree_r2}"
```

```python
# Evaluate the random forest model
forest_mae, forest_mse, forest_r2 = evaluate_model(forest_model, x_test, y_test)
f"Random Forest - MAE: {forest_mae}, MSE: {forest_mse}, R2: {forest_r2}"
```

### 4.5 Model Selection
Based on the evaluation metrics, we will select the best-performing model for our sales prediction task. The model with the lowest MAE and MSE and the highest R-squared value will be considered the best.

```python
# Compare the models and select the best one
model_performance = {
    'Linear Regression': (linear_mae, linear_mse, linear_r2),
    'Decision Tree': (tree_mae, tree_mse, tree_r2),
    'Random Forest': (forest_mae, forest_mse, forest_r2)
}

best_model_name = min(model_performance, key=lambda x: model_performance[x][1])  # Select based on lowest MSE
best_model_performance = model_performance[best_model_name]

f"Best Model: {best_model_name} - MAE: {best_model_performance[0]}, MSE: {best_model_performance[1]}, R2: {best_model_performance[2]}"
```

### 4.6 Future Sales Prediction
Using the best model, we will predict future sales and analyze the results to gain insights into expected sales trends and inventory needs.

```python
# Use the best model to predict future sales
if best_model_name == 'Linear Regression':
    best_model = linear_model
elif best_model_name == 'Decision Tree':
    best_model = tree_model
else:
    best_model = forest_model
```

### 6.3. Create Flask App
Create a Flask application to serve the model. This application will have endpoints for loading the model, making predictions, and handling other requests.

```python
from flask import Flask, request, jsonify
import joblib
import os
import pandas as pd

app = Flask(__name__)

# Load the model
# model = joblib.load('../model/best_model.pkl')

# Define the relative path to the model file
model_path = os.path.join(os.path.dirname(os.path.dirname(os.path.dirname(__file__))), 'model', 'best_model.pkl')

# Load the model
model = joblib.load(model_path)

@app.route('/')
def home():
    return "Welcome to the Sales Prediction API!"

@app.route('/predict', methods=['POST'])
def predict():
    try:
        # Get JSON data from request
        data = request.get_json()
        # Convert JSON to DataFrame
        input_data = pd.DataFrame([data])
        # Predict using the loaded model
        prediction = model.predict(input_data)
        # Return the prediction result
        return jsonify({'prediction': prediction.tolist()})
    except Exception as e:
        return jsonify({'error': str(e)})

if __name__ == '__main__':
    app.run(debug=True)
```

### 7.1. Performance Monitoring
In this section, we will monitor the performance of our trained model best_model.pkl using a new dataset (sales-07-08-24.csv). To ensure the data is in the correct format for the model, we will first preprocess the new data.

### 7.2. Preprocessing the New Data
We will preprocess the new dataset to match the input format expected by the model. This involves applying OneHotEncoding to the categorical data to generate the following features: Category_Candies, Category_Cigarettes, Category_Drinks, and Category_Others.

To avoid duplicating code, we will use a Python script (general_preprocessing_script.py) that performs the same preprocessing steps we applied during the initial Data Preparation phase.

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

def preprocess_data(input_file, output_file):
    # Load the data
    data = pd.read_csv(input_file)

    # Select the category column to be encoded
    category_column = 'Category'  # Adjust if necessary

    # Apply OneHotEncoding with the correct argument
    encoder = OneHotEncoder(drop='first', sparse_output=False)
    encoded_categories = encoder.fit_transform(data[[category_column]])

    # Create a DataFrame with the new encoded columns
    encoded_df = pd.DataFrame(encoded_categories, columns=encoder.get_feature_names_out([category_column]))

    # Merge the new encoded columns with the original DataFrame
    data = data.drop(category_column, axis=1)
    data = pd.concat([data, encoded_df], axis=1)

    # Ensure the columns align with the model
    expected_columns = ['Category_Candies', 'Category_Cigarettes', 'Category_Drinks', 'Category_Others']
    for col in expected_columns:
        if col not in data.columns:
            data[col] = 0  # Add missing columns with value 0 if they don't exist

    # Save the preprocessed DataFrame to a new CSV file
    data.to_csv(output_file, index=False)
    print(f"Preprocessing completed. File saved as {output_file}")

# Execute the preprocessing for the new file
preprocess_data('../data/sales-07-08-24.csv', '../data/preprocessed_sales-07-08-24.csv')
```
